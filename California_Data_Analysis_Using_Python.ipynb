California housing dataset has total 20640 rows and 10 columns. Each column/feature is explained in detail below. 

1. **longitude** - Longitudes are vertical lines that measure east or west of the meridian in Greenwich. A measure of how far west a house is; a higher value is farther west.
2. **latitude** - Latitudes are horizontal lines that measure distance north or south of the equator. A measure of how far north a house is; a higher value is farther north.
3. **housing_median_age** - Median age of a house within a block; a lower number indicates that it a newer building
4. **totalRooms**: Total number of rooms within a block
5. **totalBedrooms**: Total number of bedrooms within a block
6. **population**: Total number of people residing within a block
7. **households**: Total number of households, i.e a group of people residing within a home unit, for a block
8. **medianIncome**: Median income for households within a block of houses (measured in tens of thousands of US Dollars)
9. **medianHouseValue**: Median house value for households within a block (measured in US Dollars as the data given is for California,USA)
10. **oceanProximity**: Location of the house with respect to ocean/sea

```python
#6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.
#print(df.head(2))
#print(df.median())


#programming construct to calculate median value
  
def calcul_median(col_name): # Function definition
  n=len(col_name) # calculating length of the column_name
  index = n // 2 # calculate index using floor division
  if n % 2:
    return sorted(col_name)[index] # Return sorted column index
  return sum(sorted(col_name)[index - 1:index + 1]) / 2 # return sorted column and slicing (index-1 to index) /2 value as median

col_name=input("Enter column name for which you want to calculate median:") # Taking user input
if col_name=="longitude": # If to handle all columns for which calculating median is possible
  col_name=df['longitude']
  print("Median of longitude is:",calcul_median(col_name)) # function call and print median value
elif col_name=="latitude":
  col_name=df['latitude']
  print("Median of latitude is:",calcul_median(col_name))
elif col_name=='housing_median_age':
  col_name=df['housing_median_age']
  print("Median of housing_median_age is:",calcul_median(col_name))
elif col_name=='total_rooms':
  col_name=df['total_rooms']
  print("Median of total_rooms is:",calcul_median(col_name))
elif col_name=='total_bedrooms':
  col_name=df['total_bedrooms']
  print("Median of total_bedrooms is:",calcul_median(col_name))
elif col_name=='population':
  col_name=df['population']
  print("Median of population is:",calcul_median(col_name))
elif col_name=='households':
  col_name=df['households']
  print("Median of households is:",calcul_median(col_name))
elif col_name=='median_income':
  col_name=df['median_income']
  print("Median of median_incomeis:",calcul_median(col_name))
elif col_name=='median_house_value':
  col_name=df['median_house_value	']
  print("Median of median_house_value	is:",calcul_median(col_name))

else:
  print("Entered Feature to calculate median is ocean proximity and it is of DDtype:Object which is categorical data") 
  # if enetered column name is Ocean proximity this else part will handle





```

```python
#10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
# If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

import numpy as np
conditions = [
              (df['total_bedrooms'] <=10),
              (df['total_bedrooms'] >=11)&(df['total_bedrooms'] <=1000),
              (df['total_bedrooms'] > 1000)
             ] # Creating conditions based on given inst
values = ['small','medium','large'] # values to be assigned based on conditions

df['total_bedroom_size']=np.select(conditions,values) # new column is added to the dataframe 'total_bedroom_size' based on conditions and values.

df
```
